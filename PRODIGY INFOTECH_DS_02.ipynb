```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Data Generation
# Seed for reproducibility
np.random.seed(42)

# Generate a random dataset
n_samples = 1000
data = {
    'Age': np.random.randint(18, 70, size=n_samples),
    'Gender': np.random.choice(['Male', 'Female'], size=n_samples),
    'Income': np.random.randint(20000, 120000, size=n_samples),
    'Spending_Score': np.random.randint(1, 100, size=n_samples),
    'City': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], size=n_samples)
}

# Create DataFrame
df = pd.DataFrame(data)

# Introduce some missing values
df.loc[np.random.choice(df.index, size=50, replace=False), 'Income'] = np.nan
df.loc[np.random.choice(df.index, size=30, replace=False), 'Spending_Score'] = np.nan

# Step 2: Data Cleaning
# Handling missing values: Fill missing 'Income' with the median income, and 'Spending_Score' with the median score
df['Income'].fillna(df['Income'].median(), inplace=True)
df['Spending_Score'].fillna(df['Spending_Score'].median(), inplace=True)

# Check for duplicates and remove if any
df.drop_duplicates(inplace=True)

# Convert 'Gender' and 'City' to categorical data types
df['Gender'] = df['Gender'].astype('category')
df['City'] = df['City'].astype('category')

# Step 3: Exploratory Data Analysis (EDA)
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# 1. Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Distribution of Income
plt.figure(figsize=(10, 6))
sns.histplot(df['Income'], kde=True, bins=30)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()

# 3. Gender Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# 4. Relationship between Age and Spending Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Spending_Score', hue='Gender', data=df)
plt.title('Age vs. Spending Score by Gender')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.show()

# 5. Relationship between Income and Spending Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Income', y='Spending_Score', hue='City', data=df)
plt.title('Income vs. Spending Score by City')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```
